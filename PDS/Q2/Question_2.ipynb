```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DataFrame = pd.read_csv( 'StudentsPerformance.csv' )
```

```python
DataFrame
```

```python
# Pie Chart of Test Preparation Course
a = DataFrame['test preparation course'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(a, labels=a.index, autopct='%1.1f%%', colors=['blue', 'yellow'])
plt.title('Pie Chart of test preparation course')
plt.show()
```

```python
# Box Plot of Math Scores by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=DataFrame, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.title('Box Plot of Math Scores by Gender')
plt.grid(axis='y')
plt.show()
```

```python
# Bar Chart of Parental level of education types
plt.figure(figsize=(10, 6))
sns.countplot(x='parental level of education', data=DataFrame, palette='rocket', order=DataFrame['parental level of education'].value_counts().index)
plt.xticks(rotation=45, ha='right')
plt.xlabel('parental level of education')
plt.ylabel('Count')
plt.title('Bar Chart of parental level of education Types')
plt.grid(axis='y')
plt.show()
```

```python
# Histogram of reading Scores
plt.figure(figsize=(8, 6))
sns.histplot(DataFrame['reading score'], bins=20, kde=True, color='tan')
plt.xlabel('reading Score')
plt.ylabel('Frequency')
plt.title('Histogram of reading Scores')
plt.grid(True)
plt.show()
```

```python
# Scatter Plot of Writing vs. Math Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='writing score', y='math score', data=DataFrame, hue='gender', palette='mako', alpha=0.9)
plt.xlabel('writing score ')
plt.ylabel('math Score')
plt.title('Scatter Plot of Writing vs. Math Scores')
plt.grid(True)
plt.legend(title='Gender')
plt.show()
```
