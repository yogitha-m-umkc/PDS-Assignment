```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DataFrame = pd.read_csv('Frailty.csv')
DataFrame
```

```python
print("Shape:", DataFrame.shape, "\n")
DataFrame.dtypes
```

```python
DataFrame.columns=['Height', 'Weight', 'Age', 'Grip Strength','Frailty']
DataFrame.columns
DataFrame['Frailty'].replace(['Y', 'N'], [1, 0], inplace=True)
DataFrame
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Frailty', data=DataFrame)
plt.title('Frailty Distribution')
plt.show()
```
